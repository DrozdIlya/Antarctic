```python
files = glob.glob('/Users/ilyadrozd/Desktop/IO/meteo/*.nc')
files.sort()
files

# df = pd.DataFrame()
for file, year in zip(files, [2015,2016,2017,2018,2019,2020,2021,2022]):
    ds = nc.Dataset(file)
    lons = ds['longitude'][:]
    lats = ds['latitude'][:]
    time = ds['time'][:]
    u10 = ds['u10'][:].flatten()
    v10 = ds['v10'][:].flatten()
    sst = ds['sst'][:].flatten()
    t2m = ds['t2m'][:].flatten()
    d2m = ds['d2m'][:].flatten()
    msl = ds['msl'][:].flatten()
    lhf = ds['slhf'][:].flatten()
    shf = ds['sshf'][:].flatten()
    srd = ds['ssrd'][:].flatten()
    trd = ds['strd'][:].flatten()
    lsm = ds['lsm'][:].flatten()
    
    timeg, latg, long = np.meshgrid(time,lats,lons, indexing='ij' )
    
    df = pd.DataFrame(np.array([timeg.flatten(), latg.flatten(), long.flatten(),
                                u10, v10, sst, t2m, d2m, msl, lhf, shf, srd, trd, lsm]).T)

    print(len(df))
    df.to_csv(f'/Users/ilyadrozd/Desktop/IO/meteo/meteo_{year}.csv', sep=' ', header=None, index=None)

    # pd.to_datetime(time, origin='1900-01-01 00:00:00.0', unit='h')

```
